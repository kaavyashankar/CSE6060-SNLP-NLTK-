# TASK 1 - CHUNKING

Chunking is a process of extracting phrases from unstructured text, which means analyzing a sentence to identify the constituents such as Noun Groups, Verbs, verb groups, etc.., It works on top of POS tagging. It uses POS-tags as input and provides chunks as output.

```python
import nltk
from nltk.corpus import state_union
from nltk import word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
```

```python
text = open("Harrison_1891.txt")
text1 = open("Harrison_1892.txt")
data = text.read()
```

```python
sentence = nltk.sent_tokenize(data)
grammar = "NP: {<DT>?<JJ>*<NN>}"
```

```python
chunk = nltk.RegexpParser(grammar)

for i in sentence:
    words = nltk.word_tokenize(i)
    tagged = nltk.pos_tag(words)
    chunked = chunk.parse(tagged)
    print(chunked)
```

```python
chunked.draw()
```
